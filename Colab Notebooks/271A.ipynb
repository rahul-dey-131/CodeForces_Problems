```python
last = int(input())
next = last

while True:
  next += 1
  nextStr = str(next)
  for i in nextStr:
    if nextStr.count(i) > 1: continue
    else: break

print(next)
```

```python
def find_beautiful_year(y):
    while True:
        y += 1
        if len(set(str(y))) == len(str(y)):
            return y

y = int(input())
print(find_beautiful_year(y))
```
