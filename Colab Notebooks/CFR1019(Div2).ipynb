```python
from statistics import median

ts = int(input())
result = []

for t in range(ts):
  [n, k] = list(map(int, input().split()))
  s = list(map(int, input().split()))
  s0, sl = s[0], s[-1]
  s = s[1:-2]
  newArr = [s0, median(s) , sl]

  if median(newArr) <= k: result.append('YES')
  else: result.append('NO')

for r in range(len(result)):
  print(result[r])
```
