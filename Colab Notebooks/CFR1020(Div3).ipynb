```python
ts = int(input())
result = [0]*ts

for t in range(ts):
  n, k = map(int, input().split())
  a = list(map(int, input().split()))
  b = list(map(int, input().split()))
  if (len(set(b)) == 1): result[t] = k + min(a) - max(a) + 1
  else:
    maxb, maxInd = max(b), b.index(max(b))
    for i in range(len(b)):
      if b[i]>-1 and b[i]+a[i] != a[maxInd] + maxb:
        result[t] = 0
        break
      elif b[i]==-1 and a[i] > (a[maxInd] + maxb):
        result[t] = 0
        break
      elif b[i]==-1 and a[maxInd] + maxb - a[i] < k: result[t] = 1

for rst in result:
  print(rst)
```

```python
def maximize_mex(n, z):
    permutation = []
    used = [False] * n

    for i in range(z):
        permutation.append(i)
        used[i] = True

    for i in range(z+1, n):
        if not used[i]:
            permutation.append(i)
            used[i] = True

    if z < n:
        permutation.append(z)

    return permutation

ts = int(input())
result = []
for t in range(ts):
  n, z = map(int, input().split())
  permutation = maximize_mex(n, z)
  result.append(" ".join(map(str, permutation)))

for rst in result:
  print(rst)
```

```python
ts = int(input())
rsts = [0]*ts

for t in range(ts):
  n = int(input())
  mains = list(input())
  result = 0
  for i in range(n):
    s = mains.copy()
    if s[i] == "0": s[i] = "1"
    else: s[i] = "0"
    result += s.count("1")
  rsts[t] = result

for rst in rsts:
  print(rst)
```
