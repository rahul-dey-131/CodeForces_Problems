```python
ts = int(input())
results = []
for t in range(ts):
    n = int(input())
    if (n % 4): results.append("NO")
    else:
        results.append("YES")
        nums = []
        for i in range(1, n//2 + 1): nums.append(i * 2)
        for j in range(1, n//2): nums.append(j * 2 - 1)
        nums.append(sum(nums[:n//2]) - sum(nums[n//2:]))
        results.append(" ".join(map(str, nums)))

for result in results: print(result)

```
