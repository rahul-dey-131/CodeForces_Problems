```python
n = int(input())
faces = 0
for i in range(n):
  polygon = input()
  if polygon == "Tetrahedron": faces += 4
  elif polygon == "Cube": faces += 6
  elif polygon == "Octahedron": faces += 8
  elif polygon == "Dodecahedron": faces += 12
  elif polygon == "Icosahedron": faces += 20
print(faces)
```
