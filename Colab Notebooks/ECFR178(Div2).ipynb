```python
def maximize_last_k_elements(t, test_cases):
    results = []
    for n, array in test_cases:
        max_sums = []
        for k in range(1, n + 1):
            current_max = sum(array[-k:])  # Sum of the last k elements in the original array
            for i in range(n):
                temp_array = array[:-1] + [array[i]]  # Move element i to the end temporarily
                current_max = max(current_max, sum(temp_array[-k:]))
            max_sums.append(current_max)
        results.append(max_sums)
    return results

# Example Input
t = int(input())
test_cases = list(map(int, input().split()))
output = maximize_last_k_elements(t, test_cases)
print(output)
```

```python
ts = int(input())
result = []
for t in range(ts):
  [a, b, c] = list(map(int, input().split()))
  eq = (a+b+c)/3
  if (a+b+c)%3 == 0:
    gapa = max(0, eq-a)
    gapb = max(0, eq-b)
    if (gapa+gapb) == c-eq: result.append("YES")
    else: result.append("NO")
  else: result.append("NO")
for rst in result:
  print(rst)
```
