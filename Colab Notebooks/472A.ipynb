```python
def design(n):
    a, b = n//2, (n-(n//2))
    while (isPrime(a) or isPrime(b)):
        a -= 1
        b += 1
    return a, b

def isPrime(m): return True if ((m%2 != 0) and (m%3 != 0) and (m%5 != 0) and (m%7 != 0)) or m in [1,2,3,5,7] else False

n = int(input())
a, b = design(n)
print(a, b)
```
