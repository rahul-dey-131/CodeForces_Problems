```python
ts =  int(input())
result = []

for t in range(ts):
  n = input()
  rst = []
  c = 0
  while (c < len(n)):
    if (int(n[c])): rst.append(n[c] + "0"*(len(n)-c-1))
    c += 1
  result.append(len(rst))
  result.append(" ".join(rst))

for rst in result: print(rst)
```
