```python
[n,t] = list(map(int, input().split()))
qs = list(input())

for t in range(t):
  q = 0
  while (q < len(qs)-1):
    if (qs[q] == "B") and (qs[q+1] == "G"):
      qs[q] = "G"
      qs[q+1] = "B"
      q += 2
    else: q += 1

print(("").join(qs))
```
