```python
def solve_wonderful_city():
    t = int(input())
    for _ in range(t):
        n = int(input())
        h = [list(map(int, input().split())) for _ in range(n)]
        a = list(map(int, input().split()))
        b = list(map(int, input().split()))

        # Try both possibilities: we either fix row 0 or column 0
        # Since we can only increase, we'll use that to our advantage
        INF = float('inf')
        res = INF

        # Try all possible values of r0 (how many times we apply worker A to row 0)
        for r0 in range(101):  # Try small bounds to prevent timeout
            # Determine c[j] based on row 0
            c = [h[0][j] + r0 - h[j][j] for j in range(n)]
            valid = True
            cost = r0 * a[0]

            for j in range(n):
                if c[j] < 0:
                    valid = False
                    break
                cost += c[j] * b[j]

            if not valid:
                continue

            # Now calculate required r[i] for other rows
            r = [0] * n
            r[0] = r0
            for i in range(1, n):
                ri = h[0][0] + r0 - h[i][0] - c[0]
                if ri < 0:
                    valid = False
                    break
                r[i] = ri
                cost += ri * a[i]

            if not valid:
                continue

            # Check if the final grid is valid
            grid = [[h[i][j] + r[i] + c[j] for j in range(n)] for i in range(n)]
            beautiful = True
            for i in range(n):
                for j in range(n):
                    if j + 1 < n and grid[i][j] == grid[i][j + 1]:
                        beautiful = False
                    if i + 1 < n and grid[i][j] == grid[i + 1][j]:
                        beautiful = False

            if beautiful:
                res = min(res, cost)

        print(res if res != INF else -1)

solve_wonderful_city()
```

```python
def minimum_gloves(test_cases):
    results = []
    for n, k, left_gloves, right_gloves in test_cases:
        matching_pairs = [min(l, r) for l, r in zip(left_gloves, right_gloves)]

        matching_pairs.sort(reverse=True)

        guaranteed_pairs = sum(matching_pairs[:k])

        total_left = sum(left_gloves)
        total_right = sum(right_gloves)
        required_gloves = total_left + total_right - guaranteed_pairs

        results.append(required_gloves)
    return results

t = int(input())
test_cases = []
for _ in range(t):
    n, k = map(int, input().split())
    left_gloves = list(map(int, input().split()))
    right_gloves = list(map(int, input().split()))
    test_cases.append((n, k, left_gloves, right_gloves))

solutions = minimum_gloves(test_cases)
for sol in solutions:
    print(sol)
```

```python
def arrangement(n, s):
  nums = list(range(1, n + 1))
  nums.reverse()
  arranged = []
  i = len(s) - 1
  while (i >= 0):
    if s[i] == "<":
      arranged.insert(0, min(nums))
      nums.remove(min(nums))
      i -= 1
    else:
      arranged.insert(0, max(nums))
      nums.remove(max(nums))
      i -= 1

  return str(nums[0]) + " " + " ".join(map(str, arranged))

ts = int(input())
results = []
for t in range(ts):
  n = int(input())
  results.append()
```
