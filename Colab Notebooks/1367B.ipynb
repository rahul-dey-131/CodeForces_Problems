```python
ts = int(input())
results = [0] * ts

for t in range(ts):
    n = int(input())
    A = list(map(int, input().split()))
    toBeOdd, toBeEven = 0, 0

    for i in range(n):
        if i % 2 == 0:
            if A[i] % 2 != 0: toBeEven += 1
        else:
            if A[i] % 2 == 0: toBeOdd += 1
    
    if toBeEven == toBeOdd: results[t] = toBeEven
    else: results[t] = -1

for result in results: print(result)
```
