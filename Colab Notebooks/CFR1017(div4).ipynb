```python
cases = int(input())
result = []

for cs in range(cases):
  x = input()
  nums = list(map(int, input().split()))
  demo = []
  for num in nums:
    init = 0
    for n in nums:
      init += (num^n)
    demo.append(init)
  result.append(max(demo))

for res in result:
  print(res)
```

```python
cases = int(input())
results = []

for case in range(cases):
  passcode = []
  n = int(input())
  grid = []
  for x in range(n):
    row = input().split()
    grid.append(row)

  for i in range(n):
    for j in range(n):
      if grid[i][j] not in passcode:
        passcode.append(grid[i][j])

  for p in range(1, len(passcode)+1):
    if str(p) not in passcode:
      passcode.insert(0, str(p))
      break

  results.append(" ".join(passcode))

for result in results:
  print("2 1") if result == "1" else print(result)
```

```python
nums = int(input())
results = []

for num in range(nums):
  country = input().split()
  modName = ""
  for word in country:
    modName += word[0]
  results.append(modName)

for result in results:
  print(result)
```
